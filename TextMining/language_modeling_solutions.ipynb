Write a function that takes as input two words (say w_1 nad w_2) and the corpus and calculates the probability of w_2 immediately following w_1. If w_2 has never followed w_1 then it should return 0 

```python
def wordProb(w_1,w_2,text):
    fdist = nltk.FreqDist(nltk.bigrams(text))
    for words,freq in fdist.items():
        if words[0] == w_1 and words[1] == w_2:
            return freq/text.count(w_1)
    return 0.0    
```

```python
wordProb('Emma','Woodhouse',emma_text)
```

Write a function that takes as input a word and returns the word with the highest probability of following it. 

```python
def mostFreqWord(w,text):
    fdist = nltk.FreqDist(nltk.bigrams(text))
    c_words = {}
    for words,freq in fdist.items():
        if words[0] == w:
            c_words[words[1]] = freq
    if len(c_words)==0:
        print("No such word...")
        return None
    for word,freq in sorted(c_words.items(),key=lambda x:x[1],reverse=True):
        return word
```

```python
mostFreqWord('The',emma_text)
```

Write a function which when given a sentence gives its probability of occurrence

```python
from nltk.tokenize import word_tokenize
```

```python
def sentenceProbability(sent,text):
    words = word_tokenize(sent)
    prob = 1.0
    for i in range(1,len(words)):
        prob = prob*wordProb(words[i-1],words[i],text)
    return prob    
```

```python
sentence = 'Her father composed himself'
sentenceProbability(sentence,emma_text)
```

Starting with the word 'Jane' create a sentence where the next word is the word with highest probability of following the previous word. You should continue until you get '.' or the sentence length reaches 5.

```python
def createSentence(s_word,text,length=5):
    sent = [s_word]
    for i in range(length - 1):
        nxt_word = mostFreqWord(sent[-1],text)
        sent.append(nxt_word)
        if nxt_word=='.':
            return sent
    return sent    
```

```python
createSentence('Jane',emma_text)
```
